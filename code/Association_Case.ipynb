**Association Rule Learning** is a data mining techniques to find associative rules between a combination of items.<br>
<br>
First step in this collaborative filtering is same like others case to Import library that is needed for analyzing and processing the data.

```
import numpy as np
import pandas as pd

from mlxtend.preprocessing import OnehotTransactions
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

### Data Description

There was a data about market basket from Astra Data Scientist Bootcamp Participant. We need to make the best market basket that can be used to give a recommendation to a new customer. So they can buy a product more than one with a good combination.

```
data = pd.read_csv("Untitled form.csv")
data
```

```
data.shape
```

There was 24 participant who fill the data and they choose 3 items in the ecommerce like Bag, Camera, Guitar, HP, Mouse, Music Pad, Racket, Router, Soap, Watch.

### Data Normalization
We need to drop timestamp and Name of the participant because we only need the item to make a market basket.

```
data['type'] = data[['Item 1','Item 2', 'Item 3']].values.tolist()
data['type']
```

After choosing the items for dataframe, we need to make the records of the data become array that can be process for association with apriori algorithm.

```
datakelas = [['HP', 'Racket', 'Watch'],
             ['HP', 'Camera', 'Watch'],
             ['Watch', 'Camera', 'Music Pad'],
             ['Camera', 'Watch', 'Mouse'],
             ['HP', 'Watch', 'Music Pad'],
             ['Watch', 'Racket', 'Camera'],
             ['HP', 'Camera', 'Watch'],
             ['Watch', 'Camera', 'Music Pad'],
             ['Racket', 'Soap', 'Guitar'],
             ['Racket', 'Camera', 'Guitar'],
             ['Camera', 'Bag'],
             ['Music Pad', 'Guitar', 'Camera'],
             ['Camera', 'Watch'],
             ['Guitar', 'Camera', 'Music Pad'],
             ['Camera', 'Watch', 'Music Pad'],
             ['Camera', 'Racket', 'Guitar'],
             ['Guitar', 'Camera', 'Watch'],
             ['Guitar', 'Watch'],
             ['Camera', 'Watch'],
             ['Soap', 'Bag', 'Guitar'],
             ['Router', 'Bag'],
             ['Watch', 'Racket', 'Soap'],
             ['Music Pad', 'Soap', 'Watch'],
             ['Mouse', 'Camera', 'Soap']]
```

### Association Rules

The modeling is 'Association Rules Learning with Apriori Algorithm'.
<br>
**Apriori algorithm** is a process used to find frequent-itemset by doing iteration on dataset. Frequent-itemset is an indication of itemset that has a frequency of occurrence more than the minimum value that has been determined.

```
oht = OnehotTransactions()
oht_array = oht.fit(datakelas).transform(datakelas)
datafix = pd.DataFrame(oht_array, columns=oht.columns_)
datafix
```

By the data above, we know that what is an items that customer buy or not. We need it to make an association rules.

Association Rule Learning is data mining techniques to find associative rules between a combination of items. 

* **Support** from an association rule is the presentation of combinations of items in a database. 
* **Confidence** of an association rule is a measure of the accuracy of a rule that can be used to measure the strength of the relationships between items in the association rule.
* **Lift** ratio is a measure to determine the strength of the association rule that has been formed and can be useful for predicting the consequent in future data sets.


#### Market Basket Based on Support

```
frequent_itemsets = apriori(datafix, min_support=0.05, use_colnames=True)
print(frequent_itemsets)
```

By the minimum support, the most wanted items in dataset is Camera, watch and guitar with the highest support. It means that the Astra Data Scientis Bootcamp participant is like to buy **Camera, Watch and Guitar**

The **market basket for two items** with the highest support is
* camera and watch
* guitar and camera
* camera and music pad <br>
It means that customer who **buy camera is likely to buy watch, guitar and music pad.**

The **market basket for three items** with the highest support is 
* Guitar, Camera and Music
* Guitar, Camera and Racket
* Camera, Watch and HP



#### Market Basket Based on Confidence

```
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

There was 4 basket items with **confidence value = 1.00** that means it items has the high correlation with another items  to be buy together and **lift value more than 1.00** that means there was a dependent relationship between an items.

* Mouse and Camera
* HP and Watch	
* Guitar, Music Pad and Camera	
* Camera, HP and Watch

### Conclusion

Based on the Association Rules, we know that **Camera, Watch, Guitar and Music Pad** is the most popular item that costumer likely to buy.

For **Three Items** the best basket item is :
